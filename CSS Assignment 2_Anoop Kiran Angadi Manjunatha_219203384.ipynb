# World Happiness Report

```python
#World Happiness Report
website_url2 = requests.get('https://en.wikipedia.org/wiki/World_Happiness_Report#2019_report').text
```

```python
soup2 = BeautifulSoup(website_url2,'lxml')
```

```python
My_table2 = soup2.find('table',{'class':'wikitable sortable'})
```

```python
for row in My_table2.findAll('tr'):
    cells=row.findAll('td')
```

```python
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]
H=[]
I=[]
```

```python
for row in My_table2.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==9:
        A.append(cells[0].text)
        B.append(cells[1].text)
        C.append(cells[2].text)
        D.append(cells[3].text)
        E.append(cells[4].text)
        F.append(cells[5].text)
        G.append(cells[6].text)
        H.append(cells[7].text)
        I.append(cells[8].text)
```

```python
coloumn_names=[]
for row in My_table2.find_all('tr'):
    for cell in row.find_all('th'):
        coloumn_names.append(cell.text.rstrip())
```

```python
coloumn_names[1]="Country"
```

```python
world_happiness_report=pd.DataFrame(A,columns=[coloumn_names[0]])
world_happiness_report[coloumn_names[1]]=B
world_happiness_report[coloumn_names[2]]=C
world_happiness_report[coloumn_names[3]]=D
world_happiness_report[coloumn_names[4]]=E
world_happiness_report[coloumn_names[5]]=F
world_happiness_report[coloumn_names[6]]=G
world_happiness_report[coloumn_names[7]]=H
world_happiness_report[coloumn_names[8]]=I
```

```python
world_happiness_report = world_happiness_report.replace('\n','', regex=True)
#world_happiness_report.set_index('Overall rank', inplace=True)
```

```python
world_happiness_report['Country'].str.strip()
world_happiness_report['Country'] = world_happiness_report['Country'].str.lstrip()
```

```python
world_happiness_report
```

```python
all_merged_dataframe=world_happiness_report.merge(countries_continents.set_index('Country'),
                                                  how='outer',left_on='Country',right_on='Country')

merged_dataframe=world_happiness_report.merge(countries_continents.set_index('Country'),
                                              how='left',left_on='Country',right_on='Country')
```

# Data Visualization

```python
import pandas as pd
import statistics
import matplotlib.pyplot as plt
df = pd.DataFrame(
    {'Score': score,
    'Country':countryVariable,})
df.Score=pd.to_numeric(df.Score)
df.plot(x='Country', y='Score', kind="bar", color="green",figsize=(30,10),fontsize=12,title="Happiness Score of Countries")
plt.show()
```

```python
print("Top 10 Happiest Countries are :")
for i in range(0,10):
    print(df['Country'][i])
```

```python
value=0
for i in range (0,len(df)):
    value+=df['Score'][i]
print("Average Of Happiness:",(value/len(df)) )
```

```python

s=statistics.median(df['Score'])
below_median=[]
for i in range (0,len(df)):
    if df['Score'][i]<s:
        below_median.append(df['Country'][i])
print("List of countries below Median Score of:",s)
print("\n")
print(below_median)
```

```python
df_corr=merged_dataframe
```

```python
convert_dict = {
                'Score': float, 
                'GDP per capita': float,
                'Social support':float,
                'Healthy life expectancy':float,
                'Freedom to make life choices':float,
                'Generosity':float,
                'Perceptions of corruption':float
                } 
  
df_corr = df_corr.astype(convert_dict) 
correlation=df_corr.corr(method='pearson')

# plot correlation matrix
labels=['Score', 'GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']
fig = plt.figure(figsize=(10, 10))
correlation.style.background_gradient(cmap='coolwarm', axis=None)
ax = fig.add_subplot(111)
cax = ax.matshow(correlation, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = [0,6,1]
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(labels)
ax.set_yticklabels(labels)
plt.show()
```

### Interpretation:  We can see from the Correlation Matrix that every variable is postiviely correlated with each other except Generocity if its compared with GDP , Social Support and Healthy Life.
